```python
def filler(ar):
    ar1=[]
    ar2=[]
    ar3=[]
    ar4=[]
    for i in np.arange(len(ar)):
        if (ar[i]=="-") or (ar[i]=="NEW") or (ar[i]=="Opening Soon") or ((ar[i]=="Temporarily Closed")) :
            continue
        else:
            ar1=np.append(ar1,(ar[i])) #ar1 the array having valid ratings
    for i in np.arange(len(ar1)):
        ar2=np.append(ar2,(int(ar1[i][0:1]+ar1[i][2:4]))/10)  #change strins in integer
    m=np.median(ar2)
    for i in np.arange(len(ar)):
        if (ar[i]=="-") or (ar[i]=="NEW") or (ar[i]=="Opening Soon") or ((ar[i]=="Temporarily Closed")) :
            ar[i]=m
            ar3=np.append(ar3,(ar[i])) 
        else:
            ar3=np.append(ar3,(ar[i])) 
    for i in np.arange(len(ar3)):
        ar4=np.append(ar4,(int(ar3[i][0:1]+ar3[i][2:4]))/10)      
    return(ar4)  
```

```python
def outliers(df11,col):
    r=[]
    #r1=[]
    q1=df11.describe().at['25%',col]
    q3=df11.describe().at['75%',col]
    iqr=q3-q1
    low_bound=q1-1.5*iqr
    upp_bound=q3+1.5*iqr
    x=np.array(df11[col])
    p=[]
    for j in np.arange(len(x)):
        if df11[col][j] < low_bound or df11[col][j] > upp_bound:
            p.append(j)
        else:
            r.append(j)
            
   # r1=[p,r]
    return p  
```

```python
out1=outliers(df1,df1.describe().columns[0])
out2=outliers(df1,df1.describe().columns[1])
out3=outliers(df1,df1.describe().columns[2])
out4=outliers(df1,df1.describe().columns[3])
out5=outliers(df1,df1.describe().columns[4])
out6=outliers(df1,df1.describe().columns[5])
```
